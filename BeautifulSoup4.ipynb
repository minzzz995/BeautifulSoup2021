```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
response = urlopen('https://en.wikipedia.org/wiki/Main_Page')
soup = BeautifulSoup(response, 'html.parser') #html.parser를 이용해 분석한 다음 soup에 넣어줌
for anchor in soup.find_all('a'): #soup중에 find_all에서 a라는 부분을 찾아서 anchor라는 변수에 넣음 
    print(anchor.get('href', '/')) #하나씩 가져온 anchor중에 주소들을 가져와서 print
```
